```python
from llnn import network as llnn
```

```python
net = llnn.LLNetwork(input_dim=2, coeffs=(2, 2))
net.add_layer(output_dim = 2)
net.add_layer(output_dim = 2)
```

```python
import numpy as np
from matplotlib import pyplot as plt

classA = llnn.LLLayer._floor_fn(np.random.rand(100, 2) * 100 + 100)
classB = llnn.LLLayer._floor_fn(np.random.rand(100, 2) * 100)
data = np.vstack((classA, classB))
np.random.shuffle(data)

plt.scatter(classA[:,0], classA[:,1])
plt.scatter(classB[:,0], classB[:,1])
plt.show()
```

```python
# Demonstrates absence of data correlation
plt.plot(data)
plt.show()
```

```python
class_field = np.array([[0, 1], [1, 0]])
outputs = np.array([ class_field[0] if d[0] > 100 else class_field[1] for d in data ])
net.learn(data, outputs)
```
